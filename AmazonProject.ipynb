```python
driver.get('https://www.amazon.com.tr/gp/bestsellers/home/ref=zg_bs_home_sm')
```

```python
sheet_list = []
def scrape_data():
    #marka_name_location = ("xpath", "//div/a/span/div")
    sellcount_name_location = ("xpath", "//span[@class='a-size-small']")
    #price_name_location = ("xpath", "//span[contains(@class, '_cDEzb_p13n-sc-price_3mJ9Z')]")
    
    marka_name_button = driver.find_elements(*marka_name_location)
    sellcount_name_button = driver.find_elements(*sellcount_name_location)
    price_name_button = driver.find_elements(*price_name_location)
    
    for marka_name, sellcount_name, price_name in zip(marka_name_button, sellcount_name_button, price_name_button):
        Marka = marka_name.text
        Sellcount = sellcount_name.text
        Price = price_name.text
        sheet_list.append((Marka, Sellcount, Price))


scrape_data()

scroll_pause_time = 2  
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
   
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(scroll_pause_time)

    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

next_button = driver.find_element("xpath", "//a[contains(@href, '/gp/bestsellers/home/ref=zg_bs_pg_2_home')]")
next_button.click()
time.sleep(3)  

scrape_data()
```

```python
df = pd.DataFrame(sheet_list)
```

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings( action = 'ignore')
```

```python
dataset = pd.read_csv('amazon_best_sellers_thisweek.csv')
```

```python
dataset
```

```python
dataset['Price'] = dataset['Price'].str.replace(' TL', '', regex=False) 
dataset['Price'] = dataset['Price'].str.replace('.', '', regex=False)
dataset['Price'] = dataset['Price'].str.replace(',', '.', regex=False)
dataset['Price'] = dataset['Price'].astype(float)
```

```python
dataset.dtypes
```

```python
x = dataset
```

```python
x
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output= False)
encoded_x = encoder.fit_transform(x)
```

```python
encoded_x
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit_transform(encoded_x[:, [1, 2]])
```

```python
encoded_x
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans( n_clusters = 3 , random_state = 7 )
kmeans.fit(encoded_x)
```

```python
pred_result = pd.DataFrame(kmeans.predict( encoded_x),columns = ['Clusters'])
pred_result
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score
silhouettescore = silhouette_score(encoded_x , pred_result['Clusters'])
print("Silhouette Score:",silhouettescore )
```

```python
daviesbouldin = davies_bouldin_score(encoded_x, pred_result['Clusters'])
print("Davies-Bouldin Index:", daviesbouldin)
```
